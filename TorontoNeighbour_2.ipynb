```python
#import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#reading html into dataframe
req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(req.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))
neighborhood=pd.DataFrame(df[0])
```

```python
neighborhood
```

```python
#ignore cell with Not assigned borough
neighborhood_2=neighborhood[neighborhood['Borough'] != 'Not assigned']
neighborhood_2
```

```python
#find duplicated postal codes
print('number of postal codes: ' + str(neighborhood_2['Postal Code'].size))
print('number of unique postal codes: ' + str(neighborhood_2['Postal Code'].nunique()))
print('There are no duplicates in postal codes')
```

```python
#find not assigned neighborhoords
neighborhood_2.loc[neighborhood_2['Neighborhood']=='Not assigned']
print('There are not any Not assigned neighborhood')
```
